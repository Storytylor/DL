```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# Load data
df = pd.read_csv("C:/Users/cokre/Downloads/Compressed/mnist_train.csv")
X = df.drop("label", axis=1).values / 255.0

# Train-test split
split = int(0.8 * len(X))
x_train, x_test = X[:split], X[split:]

# Add noise
noise = 0.5
x_train_noisy = np.clip(x_train + noise*np.random.normal(size=x_train.shape), 0, 1)
x_test_noisy  = np.clip(x_test  + noise*np.random.normal(size=x_test.shape),  0, 1)

# Model
inp = Input(shape=(784,))
x = Dense(128, activation='relu')(inp)
x = Dense(64, activation='relu')(x)
x = Dense(128, activation='relu')(x)
out = Dense(784, activation='sigmoid')(x)

model = Model(inp, out)
model.compile(optimizer='adam', loss='binary_crossentropy')

# Train
model.fit(x_train_noisy, x_train, epochs=10, batch_size=256, validation_data=(x_test_noisy, x_test))

# Predict
decoded = model.predict(x_test_noisy)

# Plot
n = 5
plt.figure(figsize=(10,4))
for i in range(n):
    plt.subplot(3,n,i+1);     plt.imshow(x_test_noisy[i].reshape(28,28), cmap='gray'); plt.axis('off')
    plt.subplot(3,n,i+1+n);   plt.imshow(decoded[i].reshape(28,28), cmap='gray'); plt.axis('off')
    plt.subplot(3,n,i+1+2*n); plt.imshow(x_test[i].reshape(28,28), cmap='gray'); plt.axis('off')

plt.show()
```
