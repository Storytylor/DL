```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define Core Activation Functions
def sigmoid(x): return 1 / (1 + np.exp(-x))
def relu(x): return np.maximum(0, x)
def leaky_relu(x, alpha=0.01): return np.where(x > 0, x, alpha * x)

# 2. Organize Titles and Functions
x = np.linspace(-10, 10, 400)

activations = [
    ("Sigmoid Activation Function", sigmoid),
    ("Tanh Activation Function", np.tanh),
    ("ReLU Activation Function", relu),
    ("Leaky ReLU Activation Function", leaky_relu)
]

derivatives = [
    ("Sigmoid Derivative (Vanishing Gradient)", lambda x: sigmoid(x) * (1 - sigmoid(x))),
    ("Tanh Derivative", lambda x: 1 - np.tanh(x)**2),
    ("ReLU Derivative (Dying ReLU)", lambda x: np.where(x > 0, 1, 0)),
    ("Leaky ReLU Derivative", lambda x: np.where(x > 0, 1, 0.01))
]

# 3. Create a Reusable Plotting Function
def plot_grid(data_list):
    plt.figure(figsize=(12, 8))
    for i, (title, func) in enumerate(data_list, 1):
        plt.subplot(2, 2, i)
        plt.plot(x, func(x))
        plt.title(title)
        plt.grid(True)
    plt.tight_layout()
    plt.show()

# 4. Generate the Identical Output (2 Figures)
plot_grid(activations)
plot_grid(derivatives)
```
