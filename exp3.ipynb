```python
import numpy as np 
import matplotlib.pyplot as plt 

# 1. Define Inputs and Parameters for ANDNOT (x1 AND NOT x2)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
weights = np.array([1, -1])
threshold = 1

# 2. Vectorized McCulloch-Pitts Neuron
# Calculates the dot product and applies the threshold in one swoop
outputs = np.where(np.dot(X, weights) >= threshold, 1, 0)

# 3. Print Truth Table
print("X1 X2 | Output")
print("------|-------")
for (x1, x2), y in zip(X, outputs):
    print(f" {x1}  {x2} |   {y}")

# 4. Visualization
plt.figure(figsize=(6, 6))

# Plot the four logic states
plt.scatter(X[:,0], X[:,1], c=outputs, cmap='coolwarm', s=200, edgecolor='k', zorder=3)

# Plot decision boundary: w1*x1 + w2*x2 = threshold  =>  x2 = (threshold - w1*x1) / w2
x_line = np.linspace(-0.5, 1.5, 100)
y_line = (threshold - weights[0] * x_line) / weights[1]
plt.plot(x_line, y_line, 'k--', label='Decision Boundary')

# Format the chart
plt.xlabel("Input X1")
plt.ylabel("Input X2")
plt.title("ANDNOT Logic using McCulloch-Pitts Neuron")
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
